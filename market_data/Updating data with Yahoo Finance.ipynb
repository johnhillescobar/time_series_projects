```python
pd.options.display.max_rows = False
warnings.filterwarnings('ignore')
```

## Downloading Data

```python
today_date = datetime.datetime.now()
```

```python
raw_data = yfinance.download(tickers = "^GSPC ^FTSE ^N225 ^GDAXI BRK-A C",
                             start = '1994-01-07',
                             end = today_date,
                             interval = '1d',
                             group_by = 'ticker',
                             auto_adjust = True,
                             treads = True
)
```

```python
raw_data.head(3)
```

```python
raw_data.tail(3)
```

```python
# create a back up copy
df_comp = raw_data.copy()
```

```python
df_comp['spx'] = df_comp['^GSPC'].Close
df_comp['dax'] = df_comp['^GDAXI'].Close
df_comp['ftse'] = df_comp['^FTSE'].Close
df_comp['nikkei'] = df_comp['^N225'].Close
df_comp['brk_a'] = df_comp['BRK-A'].Close
df_comp['citi'] = df_comp['C'].Close
```

```python
df_comp = df_comp.iloc[1:].copy()
del df_comp['^GSPC']
del df_comp['^GDAXI']
del df_comp['^FTSE']
del df_comp['^N225']
del df_comp['BRK-A']
del df_comp['C']
df_comp = df_comp.asfreq('b')
df_comp = df_comp.fillna(method= 'ffill')
```
